```python
import os
import os.path as op
import glob
import matplotlib.pyplot as plt
import numpy as np
import nibabel as nb
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
# some, perhaps, relevant links:
# http://stats.stackexchange.com/questions/24314/when-is-a-repeated-measures-anova-preferred-over-a-mixed-effects-model
# http://nbviewer.ipython.org/urls/umich.box.com/shared/static/6tfc1e0q6jincsv5pgfa.ipynb
# http://jpktd.blogspot.jp/2013/03/multiple-comparison-and-tukey-hsd-or_25.html
# http://statsmodels.sourceforge.net/devel/anova.html

resdir = '/Users/cplanting/Projects/BERA_click_chirp'
data_bera = pd.read_csv(op.join(resdir,'ampl.csv'), sep=';',header=0)
print data_bera.dtypes

g0 = sns.factorplot("stim_level","amp","stim_type",kind="auto",
                    ci=90,n_boot=5000, estimator=np.mean, 
                    data=data_bera, size=6, palette="muted")

g1 = sns.factorplot("stim_level","lat","stim_type",kind="auto",
                    ci=90,n_boot=5000, estimator=np.mean, 
                    data=data_bera, size=6, palette="muted")
```

```python
model = sm.MixedLM.from_formula("amp ~ stim_level - stim_type", data_bera, groups=data_bera["sub_id"])
result = model.fit()
print result.summary()
```

```python
model_anova = ols('amp ~ C(stim_level)*C(stim_type)',data=data_bera).fit()
table = sm.stats.anova_lm(model_anova, typ=2) # Type 2 ANOVA DataFrame
print model_anova.summary()
print table
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
print pairwise_tukeyhsd(data_bera['amp'], data_bera['stim_level'])
print pairwise_tukeyhsd(data_bera['amp'], data_bera['stim_type'])
```
