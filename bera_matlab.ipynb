```python
#http://www.audiologieboek.nl/htm/hfd8/beelden/8-3-9-1.gif
subid = 'extract Heil Michael cl ads.xml.mat'
#'Schrijver Daan ACZ Neo.xml.mat'
matdir = '/Users/cplanting/ProjectsScience/ABR/processed_data/'
mat_contents = sio.loadmat(op.join(matdir,subid), struct_as_record=True)
tmp = sio.whosmat(op.join(matdir,subid))
#print tmp
t = mat_contents['ABR'][0,0][0].squeeze()
left = mat_contents['ABR'][0,0][1]
left_n_levels = left.shape[1]
print left_n_levels

right = mat_contents['ABR'][0,0][2]
right_n_levels = right.shape[1]
print right_n_levels
n = np.max((right_n_levels,left_n_levels))
cmap = plt.cm.get_cmap('winter',n)    #discrete colors
colors = cmap(np.linspace(0,1,n))

fig = plt.figure(figsize=(16, 5))                                                               
ax0 = fig.add_subplot(1,2,1)
for i in range(right_n_levels):
    level = right[0,i][0].squeeze()
    level_str = str(level) + ' dB'
    #print level
    ampl_r = right[0,i][6].squeeze()
    offset = level/35000.
    #plt.plot(t,ampl_r + offset,label=level_str,lw=3,color=colors[i])
    ax0.plot(t,ampl_r + offset,label=level_str,lw=3,color=colors[i])
    major_ticks = np.arange(0, 16, 5)                                              
    minor_ticks = np.arange(0, 16, 1)   
    plt.legend()
ax0.set_xticks(major_ticks)
ax0.set_xticks(minor_ticks, minor=True)
ax0.grid(which='minor', alpha=0.7,lw=1)                                                
ax0.grid(which='major', alpha=1,lw=1.5)  

ax1 = fig.add_subplot(1,2,2)
for i in range(left_n_levels):
    level = left[0,i][0].squeeze()
    level_str = str(level) + ' dB'
    #print level
    ampl_l = left[0,i][6].squeeze()
    offset = level/35000.
    #plt.plot(t,ampl_r + offset,label=level_str,lw=3,color=colors[i])
    ax1.plot(t,ampl_l + offset,label=level_str,lw=3,color=colors[i])
    major_ticks = np.arange(0, 16, 5)                                              
    minor_ticks = np.arange(0, 16, 1)   
    plt.legend()

ax1.set_xticks(major_ticks)
ax1.set_xticks(minor_ticks, minor=True)
ax1.grid(which='minor', alpha=0.7,lw=1)                                                
ax1.grid(which='major', alpha=1,lw=1.5)  
plt.xlabel('time (ms)')
plt.ylabel('amplitude (muV)')
plt.title('Clicks')
plt.savefig('MH_clicks.png')
```
